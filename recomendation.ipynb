```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer('english')
from nltk.corpus import stopwords
import string
stopword = set(stopwords.words('english'))
```

```python
def clean(text):
    print(text)
    text = str(text).lower()
    text = re.sub('\[.*?\]','',text)
    text = re.sub('https?://\S+|www\.\S+','',text)
    text = re.sub('<.*?>+','',text)
    text = re.sub('[%s]' % re.escape(string.punctuation),'',text)
    text = re.sub('\n','',text)
    text = re.sub('\w*\d\w*','',text)

    text = [word for word in text.split(' ') if word not in stopword]
    text = ''.join(text)

    print(text)
    return(text)

for ind,i in enumerate(fdata['Title']):
    fdata['Title'][ind] = clean(i)
```

```python
fdata
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
feature = fdata['Genres'].tolist()
tf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tf.fit_transform(feature)
similarity = cosine_similarity(tfidf_matrix)
indices = pd.Series(fdata.index,fdata['Title']).drop_duplicates()
```

```python
def netflix_rcmd(title):
    try:
        index = indices[title]
        similarity_scores = list(enumerate(similarity[index]))
        similarity_scores = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
        similarity_scores = similarity_scores[0:10]
        movieindices = [i[0] for i in similarity_scores]
        return fdata['Title'].iloc[movieindices]
    except:
        print("Movie Unavailable")
```

```python
print(netflix_rcmd('دفعةالقاهرة'))
```
